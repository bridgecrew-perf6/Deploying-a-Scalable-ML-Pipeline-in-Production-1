```python
def convert_to_snake_case(c):
    c = c.strip()
    c = "_".join(c.split('-'))
    return c
    

df.columns = [convert_to_snake_case(column) for column in df.columns]
```

```python
cat_features = [
    "workclass",
    "education",
    "marital_status",
    "occupation",
    "relationship",
    "race",
    "sex",
    "native_country",
]

for cat in cat_features:
    df[cat] = df[cat].str.strip()
```

```python
df.to_csv("cleaned_data.csv")
```
